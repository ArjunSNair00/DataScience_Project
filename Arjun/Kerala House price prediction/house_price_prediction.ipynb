```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['price'],axis=1)
y = data['price']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,10))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
train_data['age'] = np.log(train_data['age'] + 1)
train_data['amneties'] = np.log(train_data['amneties'] + 1)
train_data['area'] = np.log(train_data['area'] + 1)
train_data['atmDistance'] = np.log(train_data['atmDistance'] + 1)
train_data['balconies'] = np.log(train_data['balconies'] + 1)
train_data['bathrooms'] = np.log(train_data['bathrooms'] + 1)
train_data['hospitalDistance'] = np.log(train_data['hospitalDistance'] + 1)
train_data['restrauntDistance'] = np.log(train_data['restrauntDistance'] + 1)
train_data['schoolDistance'] = np.log(train_data['schoolDistance'] + 1)
train_data['shoppingDistance'] = np.log(train_data['shoppingDistance'] + 1)
train_data['status'] = np.log(train_data['status'] + 1)
```

```python
train_data.hist(figsize=(15,10))
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x='atmDistance',y='hospitalDistance',data=train_data,hue='price',palette='coolwarm')
```

```python
train_data['bathrooms_ratio'] = train_data['bathrooms'] / train_data['area']

# Rough estimate of total rooms
train_data['total_rooms_est'] = train_data['bathrooms'] + train_data['balconies'] + 1  # +1 for living room
train_data['households_est'] = 1
train_data['household_rooms'] = train_data['total_rooms_est'] / train_data['households_est']
```
